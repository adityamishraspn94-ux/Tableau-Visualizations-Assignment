# **Tableau Visualizations Assignment**
---

## **Task 1: Category Performance Analysis**

### **Chart**:

* Bar Chart (Sales by Category)

* Rows → Category
* Columns → SUM(Sales)
* Color → SUM(Profit)



* **Analysis:**

  * Technology category has the highest total sales
  * Furniture and Office Supplies have lower sales

* **Conclusion:**

  * Technology is the top-performing category

* **Business Insight:**

  * Focus more on Technology products for higher revenue
